```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Set Style
sns.set(style="darkgrid")
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'] * 25,
    'Value': np.random.randn(100),
    'Group': np.random.choice(['X', 'Y'], size=100)
})
```

```python
# 2. Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
sns.lineplot(x=x, y=y, color='blue', label='Sine Wave')
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python
# 3. Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Value', data=data, ci=None, palette='viridis')
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Average Value")
plt.show()
```

```python
# 4. Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='Value', y='Category', hue='Group', style='Group', palette='deep')
plt.title("Scatter Plot Example")
plt.xlabel("Value")
plt.ylabel("Category")
plt.show()
```

```python
# 5. Histogram (with Kernel Density Estimate - KDE)
plt.figure(figsize=(8, 5))
sns.histplot(data['Value'], kde=True, color='purple', bins=20)
plt.title("Histogram with KDE Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 6. Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Category', y='Value', data=data, palette='Set2')
plt.title("Box Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
# 7. Heatmap
# Generate a correlation matrix
matrix = data.pivot_table(values='Value', index='Category', columns='Group', aggfunc='mean')

plt.figure(figsize=(8, 5))
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Heatmap Example")
plt.show()
```

```python
# 8. Pair Plot
# Generate a small sample dataset
pair_data = sns.load_dataset('iris')  # Built-in dataset in Seaborn

sns.pairplot(pair_data, hue='species', palette='husl')
plt.show()
```

```python
# 9. Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(x='Category', y='Value', data=data, palette='muted')
plt.title("Violin Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```
